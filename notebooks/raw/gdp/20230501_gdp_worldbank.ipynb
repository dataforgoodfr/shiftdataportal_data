```python
import pandas as pd
import requests
from pandas import json_normalize
```

```python
# Define the World Bank API URL for GDP data (indicator code: NY.GDP.MKTP.CD)
url = "https://api.worldbank.org/v2/country/all/indicator/NY.GDP.MKTP.CD?format=json&per_page=20000"

# Fetch data from the World Bank API
response = requests.get(url)
data = response.json()

# Verify we have all the results
assert data[0]["total"] < data[0]["per_page"]

# Flatten the JSON data using json_normalize
gdp_data = json_normalize(data[1])
gdp_data.head()
```

```python
# Filter and clean the DataFrame
gdp_data = gdp_data[['countryiso3code', 'date', 'country.value', 'value']]
gdp_data = gdp_data.rename(columns={'countryiso3code': 'country_code', 'date': 'year', 'country.value': 'country', 'value': 'GDP'})

# Display the GDP data
gdp_data.head()
```

```python
gdp_data.to_csv("../../processed/gdp_worldbank.csv",index = False)
```
